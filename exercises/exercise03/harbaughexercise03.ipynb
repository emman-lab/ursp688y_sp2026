### RESEARCH QUESTION 

# What is the eviction rate per rental unit in each county in Maryland with the Data Provided ? 

# The goal of this was to see which counties have the highest eviction rate per rental unit they have. A table will be created which show all dwelling units and the rental units plus all the eviciton cases. The overall goal of seeing which county has the highest eviction cases and does this corrlate to just the number of dwelling units in said county. 




```python
# import dependencies 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# the goal is to seperate the data into county by evcitions. to answer the reseearch question: which maryland county had the higest eviction cases. 
# the rate will then be calcuated by dividing the number of eviction cases by the numbers of total cases 


### importing the data 
# importing the data as a pandas frame 

eviction_data = pd.read_csv("md_eviction_warrants_through_sept2025.csv")

```

```python
# remove duplicates 
unique_cases = eviction_data.drop_duplicates("CaseNumber", keep='first')
cleandata = unique_cases

# creating table of cases per county using case number which has been sorrted to remove duplicates so the cases per county will be correct 

casepivot = cleandata.groupby("County")["CaseNumber"].count()


sortable = casepivot.sort_values(ascending=False) 
sortabledf = sortable.to_frame()

# joining mdcounty owner_occupied and renter_occupied data to case pivot table 
# I got this data from NHGIS and then removed all unnecesary columns in the open office excel like program I left it only with county owner occupied 
# reading the county csv
countyclass = pd.read_csv("mdcountydata.csv") 
countyclassdf = pd.DataFrame(countyclass)


# joining df tables by county 
joineddf = pd.merge(sortabledf, countyclassdf, on="County", how="left", )

# making new column for eviction rate then dividing the number of cases in each county by the number of renters in each county 
joineddf["Eviction Rate"] = joineddf["CaseNumber"] / joineddf["renter_occupied"]

# sorting table descedning by eviction rate 
finaldf = joineddf.sort_values("Eviction Rate", ascending=False)
```

```python
# displaying table of findings 
finaldf
```

```python
#making a bar graph to show the eviction rate by county 
plt.figure(figsize=(10,6))
plt.bar(finaldf["County"], finaldf["Eviction Rate"], color="#FFA500")  
plt.xlabel("County")
plt.ylabel("Eviction Rate")
plt.title("Eviction Rate by Rental Units for Counties in Maryland from 2023 to 2025")
# I found this when I google names are to long in x axis python plot https://matplotlib.org/stable/gallery/ticks/ticklabels_rotation.html 
plt.xticks(rotation=90)
# reading the madplotlib documentation from that website Its supposed to help with overlapping with long titles 
plt.tight_layout()
plt.show()


```

```python
# Load warrants and make sure zip codes are stored as strings without decimals
warrants_df = pd.read_csv('md_eviction_warrants_through_sept2025.csv')

# Ensure zip codes are stored as strings
warrants_df['TenantZipCode'] = warrants_df['TenantZipCode'].astype('Int64').astype('string')
warrants_df['EventDate'] = pd.to_datetime(warrants_df['EventDate'])
warrants_df['EvictedDate'] = pd.to_datetime(warrants_df['EvictedDate'])
warrants_df['SourceDate'] = pd.to_datetime(warrants_df['SourceDate'])

len(warrants_df) # How many warrants are we working with?
```

```python
# Rather than geocoding 600K+ addresses, can we get only the unique ones?
geocode_input_df = exercise03.prep_warrants_for_geocoding(warrants_df)
```

```python
# The Census Geocoder API can only accept up to 10K rows at a time, so we have to break
# our dataframe into chunks

# Split into dataframes with less than 10K rows each
geocode_input_dfs = utils.chunk_dataframe(geocode_input_df, 9999)

# Save each dataframe as a CSV without a header
utils.save_dfs_to_csv(geocode_input_dfs, 'geocode_inputs', header=False)
```

```python
# Geocode addresses with the Census Geocoder (set test=True to process only one file)
census_geocode.geocode_csvs('geocode_inputs', 'geocode_outputs', test=True)
```

```python
# Recombine outputs from geocoder into a single dataframe
geocode_output_df = exercise03.combine_census_geocoded_csvs('geocode_outputs')
len(geocode_output_df)
```

```python
# Merge geocoded address back onto the inputs with separate fields for address, city, state, and zip
geocoded_df = geocode_input_df.merge(geocode_output_df, left_index=True, right_index=True)
len(geocoded_df)
```

Parquet is a file format for tabular data that efficiently stores data of many types, including a 'geoparquet' variant that stores geometries like points, lines, and polygons. You can easily save a dataframe or geodataframe to parquet with the `.to_parquet` method. The resulting file will be much smaller and load faster than more convetional formats, such as CSV or shapefile. The downside is that you won't be able to open it with Excel, ArcGIS, or other conventional desktop software.

```python
warrants_gdf.to_parquet('md_eviction_warrants_through_sept2025.geoparquet')
```
