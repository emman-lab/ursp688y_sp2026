## Research Question
For Exercise Three, I wanted to use the provided eviction data to investigate the occurrence of serial eviction filings and their spatial relationship to metro stations in Montgomery and Prince George’s counties in Maryland. Serial eviction filings can negatively impact tenants’ future ability to find housing and can create substantial financial burdens. Investigating the spatial distribution of serial eviction filings can inform interventions to address this issue. 

For my analysis I used eviction warrant data for the whole state of Maryland from 2022 through December 2024, provided by the District Court of Maryland and the Maryland Department of Housing and Community Development. I also used metro station data from Washington DC’s open data portal. To investigate the relationship, I first cleaned and filtered the eviction data for filings in Montgomery or Prince George’s counties with address matches. I then filtered the eviction data a second time to only capture addresses that had been filed against more than once. I operationalized “serial eviction filings” as addresses with more than five eviction filings between 2022 and 2024, which captures cases above the 90th decile in terms of the number of eviction filings. I then filtered Metro station data for stations in Montgomery and Prince George’s counties and created one-mile buffers around each station. The results suggest that serial eviction filings are not concentrated around metro stations. 


```python
#load data
gdf = gpd.read_parquet('md_eviction_warrants_through_sept2025.geoparquet')
```

```python
# Filter eviction filings for those in Mongomery or Prince George's Counties, those with an address match, and those with an address match in MD
eviction_filings_gdf = gdf[(gdf['EventType'] == "Petition - For Warrant of Restitution Filed") 
                         & ((gdf['County'] == "Montgomery") | (gdf['County'] == "Prince George's")) 
                         & (gdf['match_status'] == "Match")
                         & (gdf['match_address'].str.split(', ').str[2] == "MD")]
```

```python
# Drop rows where event data and case number are dulplicated
eviction_filings_gdf = eviction_filings_gdf.drop_duplicates(subset=['EventDate', 'CaseNumber'])
```

```python
# To get serial filings, keep only filings where the input address appears more than once
serial_filings_gdf = eviction_filings_gdf[eviction_filings_gdf['input_address'].duplicated(keep=False)].sort_values(by='input_address')
```

```python
# Append the gdf with a column counting how many times each input address appears
serial_filings_gdf["filing_count"] = serial_filings_gdf.groupby("input_address")["input_address"].transform("count")
```

```python
# drop duplicates on input address now that we have the number of filings against each address
serial_filings_gdf = serial_filings_gdf.drop_duplicates('input_address')

```

```python
serial_filings_gdf['filing_count'].describe(percentiles=[i/10 for i in range(1, 10)])
```

```python
# I want to focus on worst case serial filings, so I'm filting to cases above the 90th decile in terms of the numbr of eviction filings
serial_filings_gdf = serial_filings_gdf[serial_filings_gdf['filing_count'] > 5]
```

```python
# get metro station point data 
metro_stations_gdf = gpd.read_file('Metro_Stations_Regional.geojson')
```

```python
#filter for stations in MD
md_metro_stations_gdf = metro_stations_gdf[metro_stations_gdf['ADDRESS'].str.split(', ').str[2] == "MD"].to_crs(32618)
```

```python
# create buffer
def miles_to_meters(miles):
    return miles * 1609.34

threshold_dist = miles_to_meters(1) # one mile
metro_locations_buffer = md_metro_stations_gdf.buffer(threshold_dist)
metro_locations_buffer_gdf = gpd.GeoDataFrame(md_metro_stations_gdf, geometry=metro_locations_buffer, crs=32618)
```

```python
#set geo
serial_filings_gdf = serial_filings_gdf.to_crs(32618)
```

```python
# Plot metro buffers and serial eviction filings together
ax = metro_locations_buffer_gdf.plot()
serial_filings_gdf.plot(ax=ax, color='red', markersize=0.5)
```

```python
# Which serial evictions are within 1 mile of these metro staions, and which aren't
# (keep those that don't overlap with buffers around by using a left join)
serial_filings_gdf = gpd.sjoin(serial_filings_gdf, metro_locations_buffer_gdf, how='left')
serial_filings_gdf = serial_filings_gdf.rename(columns={'index_right':'metro_station'})
serial_filings_gdf['metro_station'] = serial_filings_gdf['metro_station'].fillna('none')                           
```
