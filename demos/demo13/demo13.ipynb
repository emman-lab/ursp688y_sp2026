```python
df = pd.read_csv('Townhouse_Permits_2000-Present_20260427.csv')
```

## Structured String: Geographic Coordinates

```python
df['Location']
```

```python
pattern = r"POINT\s*\(\s*([-\d\.]+)\s+([-\d\.]+)\s*\)"
df[['Longitude', 'Latitude']] = df['Location'].str.extract(pattern)
df
```
